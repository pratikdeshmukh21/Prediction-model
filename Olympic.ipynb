```python
import pandas as pd
teams = pd.read_csv("teams.csv")
print(teams)
```

```python
teams = teams [["team","country","year","athletes","age","prev_medals","medals"]]
```

```python
teams
```

```python
import seaborn as sns
sns.lmplot(x="athletes",y="medals",data=teams,fit_reg=True,ci=None)
```

```python
teams.plot.hist(y="medals")
```

```python
teams[teams.isnull().any(axis=1)]
```

```python
teams = teams.dropna()
teams
```

```python
train = teams[teams["year"]<2012].copy()
test = teams[teams["year"]>=2012].copy()
```

```python
train.shape
test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictor =["athletes","prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictor],train[target])
```

```python
LinearRegression()
prediction = reg.predict(test[predictor])
prediction
```

```python
test["prediction"]= prediction
test
```

```python
test.loc[test["prediction"]<0,"prediction"]=0
test
```

```python
test["prediction"] = test["prediction"].round()
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"],test["prediction"])
error
```

```python
errors = (test["medals"]-test["prediction"]).abs()
errors
```

```python
error_by_team = errors.groupby(test["team"])
error_by_team = error_by_team.mean()
error_by_team
```

```python
medals = test["medals"].groupby(test["team"]).mean()
medals
```

```python
error_ratio = error_by_team/medals
error_ratio
```
